Exercise 7
Find all the four-letter words in text6. With the help of a frequency distribution
(FreqDist), show these words in decreasing order of frequency.

```python
w_four = [w for w in text6 if len(w) == 4]
print(len(w_four), " words")
fdist = FreqDist(w_four)
print(fdist.most_common())
```

```python
plt.figure(figsize=(20, 20))  # the size you want
fdist.plot(50, cumulative=True)
```
